```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['^GSPC', '^VIX', '^VIX3M']
data = yf.download(tickers, start='2016-01-01')['Close']

data['Log_Ret'] = np.log(data['^GSPC'] / data['^GSPC'].shift(1))
data['RV'] = data['Log_Ret'].rolling(window=21).std() * np.sqrt(252) * 100

data['VRP_Spread'] = data['^VIX'] - data['RV']
```
